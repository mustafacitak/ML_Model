```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

print ("\n")
print(df.head())
print ("\n")
print ("\n")

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Model Doğruluk%: {accuracy * 100}%')
print ("\n")
print ("\n")

new_flower = np.array([[5.0, 3.0, 1.0, 0.2]])
predicted_species = model.predict(new_flower)
print(f'Sınıflandırılan Çiçek: {iris.target_names[predicted_species]}')

# Görsel...
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(new_flower[:, 0], new_flower[:, 1], c='red', marker='x', s=100)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset')
plt.show()
```
